```python
import pandas as pd
import numpy as np
```

```python
#import csv and creating dataframe
df = pd.read_csv('vine_table.csv')
df.head()
```

```python
# Filter to greater than 20 total votes.
df = df.loc[df['total_votes'] > 20]
df.head()
```

```python
# Create a dataframe of >= 50% helpful votes
df_votes = df.loc[df['helpful_votes'] / df['total_votes'] >= .5]
df_votes.head()
```

```python
# Create paid reviews dataframe
df_paid = df_votes[df_votes['vine'] == 'Y']
df_paid.head()
```

```python
#Create unpaid reviews dataframe
df_unpaid = df_votes[df_votes['vine'] == 'N']
df_unpaid.head()
```

```python
#Total vote counts based on vine status
df_votes.groupby('vine').count()
```

```python
#Total 5 star vote counts based on vine status
df_5stars = df_votes.loc[df_votes['star_rating'] == 5]
df_5stars.groupby('vine').count()
```

```python
#Quick percentage calculation using groupby
df_5stars.groupby('vine').count() / df_votes.groupby('vine').count()
```

```python
# Additional analyis in order to pull in more data, disregarding helpful_votes percentage
df.groupby('vine').count()
```

```python
#creation of percentages on ALL votes
df_5starsAll = df.loc[df['star_rating'] == 5]
df_5starsAll.groupby('vine').count() / df.groupby('vine').count()
```
